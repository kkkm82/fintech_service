## 3개의 데이터프레임 1개로 합치기
* join()
* merge()
* concat()

```python
# passenger + ticket
test1 = pd.concat([data_passenger, data_ticket, data_surv], axis=1)
test1
```

```python
# merge = sql에서의 join / on => 기준 컬럼 결정
# 기본이 inner join 
test2 = pd.merge(data_passenger, data_ticket, how='outer', on='PassengerId')
test2
```

```python
test3 = pd.merge(left=test2, right=data_surv, how='outer', on='PassengerId')
test3
```
