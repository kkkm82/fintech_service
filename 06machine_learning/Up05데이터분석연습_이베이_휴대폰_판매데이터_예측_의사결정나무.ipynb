### startprice가 기종별 평균 가격보다 싸야 팔린다.

1. data3에 pricebyproduct2에 있는 기종별 평균 가격 추가 / [meanprice] => 평균가격 <br>
2. startprice 컬럼과 비교해서 isUnderMean 컬럼 추가 => 평균보다 저렴하면 1, 비싸면 0

```python
def mean_price(productline):
    return pricebyproduct2.loc[pricebyproduct2.index == productline, 'mean'].values[0]

data3['meanprice'] = data3['productline'].apply(mean_price)
```

```python
# 위와 동일 코드 
# data3['productline'].apply(lambda x: pricebyproduct2.loc[x,'mean'])
```

```python
data3
```

```python
def mean_price2(row):
    if row['meanprice'] > row['startprice']:
        return 1
    elif row['meanprice'] < row['startprice']:
        return 0

data3['isUnderMean'] = data3.apply(mean_price2, axis=1)
```
