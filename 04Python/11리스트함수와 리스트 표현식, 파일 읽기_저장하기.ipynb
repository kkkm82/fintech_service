## 리스트 함수
* min(): 리스트 내에서 최소값 찾아주는 함수
* max(): 리스트 내에서 최대값을 찾아주는 함수
* sum(): 리스트 내의 요소를 더해 합계를 구해주는 함수

```python
a = [38, 21, 53, 62, 19, 5, 78, 55, 42, 33]
b = [77, 65, 48, 79, 5, 4, 35, 3, 99, 105]
```

* 정렬해서 가장 앞에 있는 숫자 추출

```python
sorted(a)[-1]
```

```python
sorted(a, reverse=True)[-1]
```

```python
# 리스트 내에서 가장 큰 값을 찾는 코드
temp = b[0]
for i in b:
    if temp <= i:
        temp = i
print("최대값:", temp )
```

```python
# 리스트 내에서 가장 큰 값을 찾는 코드
temp = b[0]
for i in b:
    if temp >= i:
        temp = i
print("최소값:", temp)
```

```python
print(min(a))
print(max(a))
print(min(b))
print(max(b))
```

```python
import time
```

```python
start = time.time()

a = []
for i in range(1, 200000000):
    a.append(i + 5)
print(a[:10])

print('코드 실행 시간: ', time.time() - start)
```

```python
start = time.time()

b = [i + 5 for i in range(1, 200000000)]

print(b[:10])

print('코드 실행 시간: ', time.time() - start)
```

## with 문으로 파일 자동으로 닫기
* 파일 만들기 순서 : 파일 만들기 => 내용 적기 => 파일닫기(저장)<br>
* 이때 close()를 안 할 시에 오류 발생<br>
* 위와 같은 실수를 줄이기 위해 파일을 자동으로 닫아주는 with문을 사용하면 편리

```python
with open('test_with.txt', 'w') as file3: # as 뒤는 파일 이름명(=별칭)
    file3.write('with open을 사용하면 파일을 자동으로 닫아주어 편리하다.')
```

```python
with open('test_with.txt', 'r') as file4: # 읽을 때는 다른 파일 이름명으로 지정
    text = file4.read()
print(text)
```

## encoding에 따른 읽기 오류

* 인코딩에 따른 오류가 발생할 수 있음

```python
# 해결 코드 2
empty_dic = {}
for fir_dic in data[:]:
    for k, v in fir_dic.items():
        empty_dic.setdefault(k, []).append(v)
print(empty_dic)
```

```python
# 해결 코드 3
result2 = {}
for key in list(data[0].keys()):
    item_list = []
    for item in data[:]:
        item_list.append(item[key])
    result2[key] = item_list
print(result2)
```

# 딕셔너리 표현식
* {} 안에 반복문을 넣어서 딕셔너리를 만드는 방법
* {key : value로 올 for문 key로 올 for문}
* {key : value for value in data key for in keys}

```python
# 해결 코드 4
result3 = {}
for key in list(data[0].keys()):
    result3[key] = [item[key] for item in data]
print(result3)
```

```python
# 해결 코드 5
result4 = {key : [item[key] for item in data] for key in list(data[0].keys())}
print(result4)
# 5와 동일 코드
# for key in list(data[0].keys()):
#     result3[key] = [item[key] for item in data]
# print(result3)
```

## if else문 1줄로 표현하기
* 참일 때 실행문 if 조건 else 거짓일 때 실행문

```python
L = list(range(20))
```

```python
for i in L:
    if i%2==1:
        print("odd")
    else:
        print("even")
```

```python
# 위와 동일 코드
for i in L:
    print("odd") if i%2==1 else print("even")
```
